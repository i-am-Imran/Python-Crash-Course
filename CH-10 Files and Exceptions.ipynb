```python
print("Give me two numbers, And i'll divide them.")
print("Enter 'q' to exit.")
while True:
    first_num  = input("First number, Or 'q' to exit: ")
    if first_num.lower() == 'q':
        break
    second_num = input("Second number, Or 'q' to exit: ")
    if second_num.lower() == 'q':
        break
        
    try:
        result = (int(first_num)/int(second_num))
        
    except ZeroDivisionError:
        print("Can't divide with zero.")
    else:
        print(result)
```

```python
def count_words(filename):
    '''Count approximate words in a file'''
    try:
        with open(filename,encoding='utf-8') as f_word:
            contents = f_word.read()
    except FileNotFoundError:
        pass
        #print(f"Sorry, the file '{filename}' is not found.")
    else:
        num_words = len(contents.split())
        print(f"The file {filename} has approximately {num_words} words.")
        
```

```python
import json
def greet_user():
    """Greet the user by name."""
    filename = 'username.json'
    try:
        with open(filename) as f:
            username = json.load(f)
    except FileNotFoundError:
        username = input("What is your name? ")
        with open(filename, 'w') as f:
            json.dump(username, f)
            print(f"We'll remember you when you come back, {username}!")
    else:
        print(f"Welcome back, {username}!")
        greet_user()
```
